```python
# Generate two dictionaries with 50 country names as the keys and random floats as the values.
# The first dict has values between 2 and 6, the second between 0 and 1
import random

country_names = ["Country" + str(i) for i in range(1, 51)]

within_community_degree = {country: random.uniform(-2, 6) for country in country_names}
participant_coefficient = {country: random.uniform(0, 1) for country in country_names}
```

```python
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use(
    "https://raw.githubusercontent.com/allfed/ALLFED-matplotlib-style-sheet/main/ALLFED.mplstyle"
)


def create_scatterplot(within_community_degree, participant_coefficient):
    assert set(within_community_degree.keys()) == set(participant_coefficient.keys()), "Both dictionaries must have the same keys"
    # check that all within_community_degree values are between -5 and 10
    assert all(-5 <= value <= 10 for value in within_community_degree.values())
    # check that all participant_coefficient values are between 0 and 1
    assert all(0 <= value <= 1 for value in participant_coefficient.values())

    # Create a data frame from the dictionaries
    df = pd.DataFrame({'within_community_degree': within_community_degree, 'participant_coefficient': participant_coefficient})
    # Create a scatterplot from the data frame
    ax = df.plot.scatter(x='participant_coefficient', y='within_community_degree', zorder=10, color="black")

    ax.set_xlabel('Participant Coefficient')
    ax.set_ylabel('Within Community Degree')

    fontsize = 8

    # Color the background of the plot using ax.fill_between
    # Provincial hubs have within_community_degree between 1 and 6 and participant_coefficient between 0 and 0.3
    ax.fill_between([0, 0.3], [1, 1], [6, 6], color='red', alpha=0.3)
    # Add a label for the provincial hubs
    ax.text(0.01, 5.5, 'Provincial hub', fontsize=fontsize, color='red', alpha=0.5)

    # Connector hubs have within_community_degree between 1 and 6 and participant_coefficient between 0.3 and 0.75
    ax.fill_between([0.3, 0.75], [1, 1], [6, 6], color='green', alpha=0.3)
    # Add a label for the connector hubs
    ax.text(0.31, 5.5, 'Connector hub', fontsize=fontsize, color='green', alpha=0.5)

    # Kinless hubs have within_community_degree between 1 and 6 and participant_coefficient between 0.75 and 1
    ax.fill_between([0.75, 1], [1, 1], [6, 6], color='gold', alpha=0.3)
    # Add a label for the kinless hubs
    ax.text(0.76, 5.5, 'Kinless hub', fontsize=fontsize, color='gold', alpha=1)

    # Ultra-peripheral non-hubs have within_community_degree between 1 and -2 and participant_coefficient between 0 and 0.05
    ax.fill_between([0, 0.05], [-2, -2], [1, 1], color='purple', alpha=0.3)
    # Add a label for the ultra-peripheral hubs
    ax.text(0.001, -1.8, 'Ultra\nperi-\npheral\nnon\nhub', fontsize=fontsize, color='purple', alpha=0.5)

    # Peripheral non-hubs have within_community_degree between 1 and -2 and participant_coefficient between 0.05 and 0.62
    ax.fill_between([0.05, 0.62], [-2, -2], [1, 1], color='blue', alpha=0.3)
    # Add a label for the peripheral non-hubs
    ax.text(0.06, -1.8, 'Peripheral non-hub', fontsize=fontsize, color='blue', alpha=0.5)

    # Connector non-hubs have within_community_degree between 1 and -2 and participant_coefficient between 0.62 and 0.8
    ax.fill_between([0.62, 0.8], [-2, -2], [1, 1], color='orange', alpha=0.3)
    # Add a label for the connector non-hubs
    ax.text(0.63, -1.8, 'Connector non-hub', fontsize=fontsize, color='orange', alpha=1)

    # Kinless non-hubs have within_community_degree between 1 and -2 and participant_coefficient between 0.8 and 1
    ax.fill_between([0.8, 1], [-2, -2], [1, 1], color='brown', alpha=0.3)
    # Add a label for the kinless non-hubs
    ax.text(0.81, -1.8, 'Kinless non-hub', fontsize=fontsize, color='brown', alpha=1)


create_scatterplot(within_community_degree, participant_coefficient)
```
